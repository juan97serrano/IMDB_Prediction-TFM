```
data_tab['ACTOR_0'] = data_tab['ACTOR_0'].apply(lambda x: x.replace(' ',''))
data_tab['ACTOR_1'] = data_tab['ACTOR_0'].apply(lambda x: x.replace(' ',''))
data_tab['ACTOR_2'] = data_tab['ACTOR_2'].apply(lambda x: x.replace(' ',''))
```

```
def listWithTheBest(data_tab,columnName,number=3):

  data =  data_tab[columnName].value_counts()>number
  data = data.reset_index()

  final_list = data.loc[data[columnName]==True]
  final_list = final_list['index'].values
  return final_list
```

```
def new_valuesInDataframe(data_tab,listWithTeBest,columnName,nuevoNombre):
  for i, row in data_tab[columnName].iteritems():
    if row not in listWithTeBest:
      data_tab.at[i,columnName]=nuevoNombre
  return data_tab
```

```
best_writers = listWithTheBest(data_tab,'ACTOR_0',3)
data_tab = new_valuesInDataframe(data_tab,best_writers,'ACTOR_0',"PepitoDoe")

best_directors = listWithTheBest(data_tab,'ACTOR_1',3)
data_tab = new_valuesInDataframe(data_tab,best_directors,'ACTOR_1',"PepitaDoe")

best_directors = listWithTheBest(data_tab,'ACTOR_2',3)
data_tab = new_valuesInDataframe(data_tab,best_directors,'ACTOR_2',"JosefinaDoe")
```
